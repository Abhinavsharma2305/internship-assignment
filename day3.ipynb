```python
# question 1 
# Dictionary Operations

student = {"name" : "Rahul", "age" : 20, "course" : "BCA"}

# print dictionary student 
print(student)

# access value from the student dictionary
print(student["name"])

# add new item to studetn dictionary
student["city"] = "Jaipur"

# update value In student dictionary
student["age"] = 21

print("\nafter update")
print(student)

# length of dictionary student dictionary
print("\ntotal items =", len(student))

# print keys of the student dictionary
print("\nkeys are")
for i in student.keys():
    print(i)

# print values of the student dictionary
print("\nvalues are")
for i in student.values():
    print(i)

# remove item from the student dictionary
student.pop("course")

print("\nafter removing")
print(student)


```

```python
# Tuple Operations 

numbers = (10,20,30,40,50)

# print tuple
print(numbers)

# access values
print("first value =", numbers[0])
print("last value =", numbers[-1])

# length
print("length =", len(numbers))

# count
print("count of 20 =", numbers.count(20))

# index
print("index of 40 =", numbers.index(40))

# loop
print("\nall values")
for i in numbers:
    print(i)

# checking value
if 30 in numbers:
    print("30 is present")
```

```python
# Set Operations

A = {1,2,3,4}
B = {3,4,5,6}

# print sets
print("set A =", A)
print("set B =", B)

# add item
A.add(7)

print("\nafter adding")
print(A)

# remove item
A.remove(2)

print("\nafter removing")
print(A)

# loop
print("\nlooping set")
for i in A:
    print(i)

# union
print("\nunion =", A.union(B))

# intersection
print("intersection =", A.intersection(B))

# difference
print("difference =", A.difference(B))

```

```python
# caluculations using functions 

# using addition function
def add(a, b):
    return a + b

# using subtraction function
def sub(a, b):
    return a - b

# using multiplication function
def multiply(a, b):
    return a * b

# using division function
def divide(a, b):
    return a / b


# taking input from user
num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))

# calling functions
print("\nAddition =", add(num1, num2))
print("Subtraction =", sub(num1, num2))
print("Multiplication =", multiply(num1, num2))

# checking divide by zero
if num2 != 0:
    print("Division =", divide(num1, num2))
else:
    print("Division not possible")
```

```python
# palindrome number program

num = int(input("Enter a number : "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reverse number is :", reverse)

if original == reverse:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```
