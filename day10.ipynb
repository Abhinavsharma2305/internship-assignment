```python
#q1
import pandas as pd
import numpy as np
a = np.array([[9, -6, 12, -210], [np.nan, -14, 90, 92]])
print("Array")
print(a)
df = pd.DataFrame(a)
print("\nAfter replacing nan with Zero\n")
print(df.replace({np.nan:0}))
arr = np.resize(df, (3,3))
print("\nIn 3*3 Size\n",arr)
arr = arr.T
print("\ntranspose\n\n",arr)
```
