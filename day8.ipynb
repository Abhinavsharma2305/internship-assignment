```python
import pandas as pd
df1 = pd.DataFrame({
    "Id": [1, 2, 3, 4],
    "name": ["B1", "B2", "B3", "B4"],
    "subject_id": ["sub1", "sub2", "sub3", "sub4"]
})


df2 = pd.DataFrame({
    "Id": [1, 2, 3, 4, 5, 7],
    "Name": ["A1", "A2", "A3", "A4", "A5", "A77"],
    "subjects_id": ["Subj8", "Subj3", "Subj9", "Subj99", "Subject33", "Subj323"]
})

print("DF1")
print(df1)

print("DF2")
print(df2)

inner_merge = pd.merge(df1, df2, on="Id", how="inner")

print("\nInner Merge Result")
print(inner_merge)
```

```python
res = df1.merge(df2 , on="Id" , how = "left")
print(res)
```

```python
res = df1.merge(df2 , on="Id" , how = "right")
print(res)
```

```python
left = df1.set_index("Id")
right = df2.set_index("Id")

index_join = left.join(
    right,
    lsuffix="_left",
    rsuffix="_right"
)

print(index_join)
```

```python
df1 = pd.DataFrame({
    "Id": [1, 2, 3, 4],
    "name": ["B1", "B2", "B3", "B4"],
    "subject_id": ["sub1", "sub2", "sub3", "sub4"]
})

df2 = pd.DataFrame({
    "Id": [1, 2, 3, 4, 5, 7],
    "Name": ["A1", "A2", "A3", "A4", "A5", "A77"],
    "subjects_id": ["Subj8", "Subj3", "Subj9", "Subj99", "Subject33", "Subj323"]
})

right_join = pd.merge(df1, df2, on="Id", how="right")

print("Right Join Result:")
print(right_join)



left = df1.set_index("Id")
right = df2.set_index("Id")

index_join = left.join(
    right,
    lsuffix="_left",
    rsuffix="_right"
)

print("\nIndex Based Join:")
print(index_join)
```

```python
df1 = pd.DataFrame({
    "Id": [1, 2],
    "Name": ["A1", "A2"]
})

df2 = pd.DataFrame({
    "Id": [3, 4],
    "Name": ["A3", "A4"]
})


df3 = pd.DataFrame({
    "Id": [1, 2, 3, 4],
    "Subject": ["Math", "Science", "English", "Physics"]
})

concat_df = pd.concat([df1, df2], axis=0)

print("Concatenated DataFrame:")
print(concat_df)

final_df = pd.merge(concat_df, df3, on="Id")

print("\nMerged DataFrame:")
print(final_df)
```
