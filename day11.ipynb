```python
#q1
import numpy as np
arr1 = np.array([9, 8, 7, 5])
arr2 = np.array([[2, 1], [4, 6]])
res = arr1.reshape(2,2)
result = np.concatenate((res, arr2), axis=0)
print("Result of concatenation along axis 0:\n", result)
```

```python
#q4
arrg = np.array([[10, 20, 30],
                [40, 50, 60]])

print("Element at row 1, column 2:", arrg[1, 2])
total = 0
rows, cols = arrg.shape
for row in arrg:
    for value in row:
        total += value
print("Rows:", rows)
print("Columns:", cols)
print("Array:")
print(arrg)
print("Sum =", total)
```

```python
#q6
arr1 = np.array([90 , 91]) 
arr2 = np.array([89 , 99])
avg = (arr1 + arr2) / 2
print("Average of NumPy arrays:\n", avg)
print("Mean of the both array is  : " , avg.mean())
```

```python
#q6
import pandas as pd
arr1 = np.array([
    [[41 , 42], [43, 44]],
    [[45, 46], [47, 48]]
])
arr2 = np.array([
    [[29, 30], [31, 32]],
    [[33, 34], [35, 36]]
])
avg = (arr1 + arr2) / 2

mode = pd.Series(avg.reshape(-1)).mode()
print("Average of NumPy arrays:\n", avg)
print("\nMean of avg\n",np.mean(avg))
print("\n Median of avg\n",np.median(avg))
print("\nMode of avg\n",mode)
```
