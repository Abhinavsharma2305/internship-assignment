```python
#question 3.f


data = {
    "Name": ["Abhinav", "Rahul"],
    "Age": [20, 21]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)


new_row = pd.DataFrame({
    "Name": ["Aman"],
    "Age": [22]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print("\nUpdated DataFrame:")
print(df)
```
