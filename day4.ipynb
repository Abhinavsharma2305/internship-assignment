```python
def maximumnumber(a, b, c):
    if(a > b and a > c):
        print(a)
    elif(b > a and b > c):
        print(b)
    else:
        print(c)

maximumnumber(2, 3, 4)
```

```python
def multiplylist(lst):
    prod = 1

    for i in lst:
        prod = prod * i

    return prod

print(multiplylist([10 , 2, 23,45]))
```

```python
def facto(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact

print(facto(10))
```

```python
def reverse_string(str):
    rever = ""

    for i in range(len(str)-1, -1, -1):
        rever = rever + str[i]

    print(rever)

reverse_string("Programming in python")
```

```python
def check_range(n, s, e):
    if n in range(s, e + 1):
        print("Number is within the range")
    else:
        print("Number is outside the range")

check_range(6, 4, 8)
```

```python
def even_numbers(n):
    for i in n:
        if i % 2 == 0:
            print(i)

even_numbers([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
def prime(num):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                print("Not Prime")
                break
        else:
            print("Prime")
    else:
        print("Not Prime")

prime(2)
```

```python
def count_case(str):
    toupper = 0
    tolower = 0

    for i in str:
        if i.isupper():
            toupper += 1
        elif i.islower():
            tolower += 1

    print("Upper Case Letters:", toupper)
    print("Lower Case Letters:", tolower)

count_case("Hello Python")
```

```python
def distinct(lst):
    new_list = []

    for i in lst:
        if i not in new_list:
            new_list.append(i)

    print(new_list)

distinct([1,1,1,3, 2, 7,2, 3,8, 4, 4, 5])
```
