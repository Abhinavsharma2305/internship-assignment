```python
# question 1 ==> 
name = input("enter the name :")
classs = input("enter the class:")

sum = 0
for i in range(1,6):
    marks = int(input(f"Enter the marks of {i} subject:"))
    sum = sum + marks


percentage = sum/5
print(f"Name: {name}\nClass: {classs}\nTotal Marks: {sum}\nPercentage: {percentage}%")
```

```python
# question 2 ==>
str1 = input("enter the string number 1 :")
str2 = input("enter the string number 2 :")

conc = str1 + " " + str2
print(f" concenated string is : {conc}")

# applying the string methods
print(f"to lower case: {conc.lower()}")
print(f"to upper case: {conc.upper()}")
print(f"to title case: {conc.title()}")
print(f"swapcase: {conc.swapcase()}")
print(f"to capitalize: {conc.capitalize()}")
print(f"capitalize: {conc.capitalize()}")
print(f"casefold: {conc.casefold()}")
print(f"center: {conc.center(50)}")
print(f"count := {conc.count('s')}")
print(f"endswith := {conc.endswith('t')}")
print(f"find := {conc.find('s')}")
print(f"isalnum() := {conc.isalnum()}")
print(f"isdigit() := {conc.isdigit()}")
print(f"isnumeric() := {conc.isnumeric()}")
print(f"isspace(:= {conc.isspace()}")
print(f"replace := {conc.replace('s', 't')}")
```

```python
#question 4
a = int(input("Enter the first number a: "))
b = int(input("Enter the second number b: "))

# Assignment operator practice

a += b
print(f"The value of a after addition is: {a}")

a -= b
print(f"The value of a after subtraction is: {a}")

a *= b
print(f"The value of a after multiplication is: {a}")

a /= b
print(f"The value of a after division is: {a}")

a %= b
print(f"The value of a after modulus is: {a}")
```

```python
# question 5

name = input("enter the name :")
classs = input("enter the class:")

sum = 0
for i in range(1,6):
    marks = int(input(f"Enter the marks of {i} subject:"))
    sum = sum + marks


percentage = sum/5

if percentage >= 60:
    grade = "A"

elif percentage >= 50:
    grade = "B"

elif percentage >= 40:
    grade = "C"

elif percentage >= 33:
    grade = "D"

else:
    grade = "Fail"

print("Name :", name)
print("Class :", classs)
print("Total Marks :", sum)
print("Percentage :", percentage)
print("Grade :", grade)
```
