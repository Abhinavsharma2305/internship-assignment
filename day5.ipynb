```python
import sqlite3

conn = sqlite3.connect("student.db")

conn.execute("CREATE TABLE students(id INTEGER, name TEXT, age INTEGER)")

print("Table Created")

conn.execute("INSERT INTO students VALUES(1,'Abhinav',20)")
conn.execute("INSERT INTO students VALUES(2,'Abhay',21)")
conn.execute("INSERT INTO students VALUES(3,'Aditya',22)")
conn.execute("INSERT INTO students VALUES(4,'aaysuh',22)")

conn.commit()

print("Records Inserted")

data = conn.execute("SELECT * FROM students")

print("\nStudent Data")
for row in data:
    print(row)

conn.execute("UPDATE students SET age = 23 WHERE id = 1")

conn.commit()

print("\nAfter Update")

data = conn.execute("SELECT * FROM students")

for row in data:
    print(row)

conn.execute("DELETE FROM students WHERE id = 3")


conn.commit()

print("\nAfter Delete")

data = conn.execute("SELECT * FROM students")

for row in data:
    print(row)

conn.close()

print("\nDatabase Closed")
```

```python
import csv

data = []

n = int(input("Insert the number of rows you want to insert in the file : "))

for i in range(n):
    print(f"\nEnter details for person number {i+1}")

    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile Number: ")
    email = input("Enter Email: ")

    data.append([name, address, mobile, email])

with open("address_book.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Address", "Mobile", "Email"])

    writer.writerows(data)

print("\nData saved successfully")
```
